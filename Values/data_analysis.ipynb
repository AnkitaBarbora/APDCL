# cumulative sum of LOAD according to DTR No

```python
# df['cumsum_sales'] = df.groupby(['store'])['sales'].cumsum()
```

```python
c_dtr['cumulative sum'] = c_dtr.groupby(['Dtr No'])['Load'].cumsum()
```

```python
c_dtr
```

```python
c_dtr.head(75)
```

```python
c_dtr.to_csv("c_dtr_with_cumulative_sum.csv")
```

```python
# this column gives the total cumulative sum of each unique Dtr No
c_dtr['dup_cum_sum'] = c_dtr.groupby('Dtr No')['cumulative sum'].transform('max')
```

```python
#del c_dtr["dep_cum_sum"]
```

```python
c_dtr.head(75)
```

## creating another data Frame with "Dtr no" and "Cumulative sum"

```python
data = [c_dtr["Dtr No"], c_dtr["dup_cum_sum"]]

headers = ["Dtr No2", "Total Load"]

cum_sum = pd.concat(data, axis=1, keys=headers)
```

```python
cum_sum
```

```python
cum_sum.drop_duplicates(subset=["Dtr No2"], inplace=True, ignore_index= True)
```

```python
cum_sum
```

```python
cum_sum.to_csv("c_dtr_sum.csv")
```
